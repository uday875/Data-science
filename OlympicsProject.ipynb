```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Load Dataset

athletes = pd.read_csv('C:/Users/HP/Downloads/Olympics/athlete_events.csv')
region = pd.read_csv('C:/Users/HP/Downloads/Olympics/noc_regions.csv')
```

```python
# Join the dataframe

athletes_df = athletes.merge(region, how = 'left', on = 'NOC')
```

```python
# Top countries participating

Top_10_countries = athletes_df.Team.value_counts().sort_values(ascending=False).head(10)
#TEAMES = athletes_df.Team.value_counts().sort_values(ascending=False).head(10).Team()

Top_10_countries
#TEAMES
```

```python
# plot for top 10 countries

plt.figure(figsize=(12,6))
plt.title('Top 10 contries participation')
sns.barplot(x=Top_10_countries.index,y=Top_10_countries,palette='Set2')
#plt.show()
```

```python
# Age Distribution of the participation

plt.figure(figsize=(12,6))
plt.title('Age distribution of the athletes')
plt.xlabel('Age')
plt.ylabel('No of participation')
plt.hist(athletes_df.Age,bins=np.arange(10,80,2),color="green",edgecolor="red")
```

```python
# Winter Sports

winter_sports = athletes_df[athletes_df.Season == "Winter"].Sport.unique()
print(winter_sports)
#winter_sports.sum()
```

```python
# Summer sports

summer_sports = athletes_df[athletes_df.Season == "Summer"].Sport.unique()
summer_sports
```

```python
# Male and female participation

gender_count = athletes_df.Sex.value_counts()
gender_count
```

```python
# pie plot for the male and female athletes

plt.figure(figsize=(12,6))
plt.title("Gender Distribution")
plt.pie(gender_count,labels=gender_count.index,autopct="%1.1f%%", startangle=180,shadow=True)
```

```python
# Total Medals

athletes_df.Medal.value_counts()
```

```python
# Total no of female athletes in each olympics

female_participants = athletes_df[(athletes_df.Sex=='F')&(athletes_df.Season=="Summer")][['Sex','Year']]
female_participants = female_participants.groupby('Year').count().reset_index()
female_participants.head(10)
```

```python
womenolympics = athletes_df[(athletes_df.Sex=='F')&(athletes_df.Season=="Summer")]

sns.set(style='darkgrid')
plt.figure(figsize=(20,10))
sns.countplot(x='Year',data=womenolympics,palette="Spectral")
plt.title("Women Participants")
```

```python
part = womenolympics.groupby('Year')['Sex'].value_counts()

plt.figure(figsize=(20,10))
part.loc[:,'F'].plot(marker="*",c="r",lw=2,ms=20,mfc='g',mec='w')
plt.title("Plot of Female athletes over time")
```

```python
# Gold medal athletes

goldMedals = athletes_df[(athletes_df.Medal == 'Gold')]
goldMedals
```

```python
# Take only the values that are different from NaN

goldMedals = goldMedals[np.isfinite(goldMedals['Age'])]
```

```python
# Gold medal beyand 60

goldMedals['ID'][goldMedals['Age']>60].count()
```

```python
sporting_event = goldMedals['Sport'][goldMedals['Age']>60]
sporting_event
```

```python
# plot for sporting_event

plt.figure(figsize=(10,5))
plt.tight_layout()
sns.countplot(sporting_event)
plt.title("Gold Medals for Athletes over 60 Years")
```

```python
# gold medals from each country

goldMedals.Region.value_counts().reset_index(name = 'Medal').head(10)
```

```python
totalGoldMedals = goldMedals.Region.value_counts().reset_index(name = 'Medal').head(6)
g = sns.catplot(x="index",y='Medal',data=totalGoldMedals,height=5,kind='bar',palette='rocket')
g.despine(left=True)
g.set_xlabels('Top 5 Contries')
g.set_ylabels('No of Medals')
#plt.figure(figsize=(20,10))
plt.title('Gold Medals for countries')
```

```python
max_year = athletes_df.Year.max()
max_year

team_name = athletes_df[(athletes_df.Year == max_year) & (athletes_df.Medal == 'Gold')].Team
team_name.value_counts().head(10)
```

```python
sns.barplot(x=team_name.value_counts().head(20),y=team_name.value_counts().head(20).index)
plt.ylabel(None)
plt.xlabel("Countrywise Gold Medals for the year 2016")
```

```python

not_null_medals = athletes_df[(athletes_df['Height'].notnull()) & (athletes_df['Weight'].notnull())]

plt.figure(figsize=(12,10))
axis = sns.scatterplot(x="Height" ,y="Weight" ,data = not_null_medals ,hue='Sex',s=120,marker="*")
plt.title("Height vs Weight of olympic Medalists ")
```
